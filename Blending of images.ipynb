```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
```

```python
def plot_rgb(img):
    plt.figure(figsize=(12,10))
    img = img.astype(np.uint8)
    return plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
park = cv2.imread("/content/park.jpeg")
plot_rgb(park);
```

```python
doggy = cv2.imread("/content/doggy.jpg")
plot_rgb(doggy);
```

```python
park_copy = park.copy()
```

```python
park_copy[0:doggy.shape[0],100:100+doggy.shape[1]] = doggy
```

```python
plot_rgb(park_copy);
```

```python
height, width = park.shape[:2]
doggy = cv2.resize(doggy, (width//3, height//3))
```

```python
park_copy = park.copy()
y_offset = int(height * 0.7)
x_offset = int(width * 0.7)
y1, y2 = y_offset, y_offset + doggy.shape[0]
x1, x2 = x_offset, x_offset + doggy.shape[1]
doggy = cv2.resize(doggy, (243, 162))
park_copy[y1:y2, x1:x2] = doggy
```

```python
plot_rgb(park_copy);
```
